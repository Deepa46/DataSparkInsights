```python
# Merge sales with customer on CustomerKey
merged_df = pd.merge(sales, customer, on='CustomerKey', how='inner')

# Merge the result with product on ProductKey
merged_df = pd.merge(merged_df, product, on='ProductKey', how='inner')
```

```python
merged_df.info()
```

```python
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])
merged_df['Delivery Date'] = pd.to_datetime(merged_df['Delivery Date'])
merged_df['Birthday'] = pd.to_datetime(merged_df['Birthday'])
```

```python
#calculate total Sales Amount
merged_df["TotalSalesAmount"] = merged_df["Unit Price USD"] * merged_df["Quantity"]
merged_df[["Unit Price USD", "Quantity", "TotalSalesAmount"]]
```

```python
#calculate total Cost Amount
merged_df["TotalCost"] = merged_df["Unit Cost USD"] * merged_df["Quantity"]
merged_df[["Unit Cost USD", "Quantity", "TotalCost"]]
```

```python
# finding the profit
merged_df["Profit"] = merged_df["TotalSalesAmount"] - merged_df["TotalCost"]
merged_df[["TotalSalesAmount", "TotalCost", "Profit"]]
```
